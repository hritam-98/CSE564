```python
import pandas as pd
```

```python
data = pd.read_csv('players_22.csv')
display(data)
```

```python
#display(data.loc[data["Wage"]== '0'])
```

```python
df = data.loc[:, (data != '0').all(axis=0)].dropna()
display(df)
```

```python
#final_data = df.iloc[:600]
final_data = df[['age','overall','wage_eur','height_cm','weight_kg','pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic', 'attacking_crossing', 'attacking_heading_accuracy', 'attacking_short_passing', 'skill_dribbling', 'skill_curve', 'power_shot_power', 'power_jumping', 'mentality_aggression', 'mentality_interceptions', 'mentality_vision', 'mentality_penalties']]
```

```python
display(final_data)
```

```python
options = ['Argentina', 'Brazil', 'Germany', 'Portugal', 'Spain', 'France', 'Netherlands', 'England', 'Belgium', 'Italy']
rslt_df = final_data[final_data['nationality_name'].isin(options)]  
display(rslt_df)
```

```python
n = rslt_df.nunique(axis=0)
print(n)
```

```python
rslt_df = rslt_df.sample(n=550)
display(rslt_df)
```

```python
rslt_df.to_csv('final_data.csv')
```
